```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

data = '/Users/pranav/nba_allNBA_predictor/final_data.csv'
```

```python
df = pd.read_csv(data)
```

```python
check = ['Pos', 'Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P',
       '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'TM_WIN%',
       'TM_SRS']
for column in check:
    fig, ax = plt.subplots(figsize=(5, 2.5))
    if column == 'Pos':
        sns.barplot(df[column].value_counts().index, df[column].value_counts())
        plt.title('Distribution of {}'.format(column))
        plt.xlabel(column)
        plt.ylabel('freq')
        plt.show()
    else:
        sns.distplot(df[column])
        plt.title('Distribution of {}'.format(column))
        plt.show()
        print ("Skewness of {}: ".format(column) + str(df[column].skew()))
        print ("Kurtosis of {}: ".format(column) + str(df[column].kurtosis()) + "\n")
```

```python
temp = ['G', 'GS', 'MP', 'eFG%', 'PTS', 'TM_WIN%', 'TM_SRS']

df_temp = df[temp]

sns.heatmap(df_temp.corr(), annot=True)
```

```python
arr = ['G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P','3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB',
       'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'TM_WIN%',
       'TM_SRS']
for stat in arr:
    co = round(df[stat].corr(df['allNBA']), 2)
    print("The correlation between {} and All-NBA success is {}".format(stat, co))
```
